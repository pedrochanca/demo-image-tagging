## Load Images

Load Counter-Strike 2 Weapon skins images in .jpeg format

```python
import requests
from bs4 import BeautifulSoup
import json
import time
import random


def scrape_weapon_skins(weapon_name):
    weapon_url = weapon_name.lower()
    base_url = f"https://csgoskins.gg/weapons/{weapon_url}"

    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36",
        "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8",
    }

    try:
        time.sleep(random.uniform(1, 2))
        response = requests.get(base_url, headers=headers, timeout=10)
        response.raise_for_status()

        soup = BeautifulSoup(response.text, "html.parser")
        json_scripts = soup.find_all("script", type="application/ld+json")

        skins = {}

        data = json.loads(json_scripts[0].string)

        print(data)

        list_of_items = data["itemListElement"]

        for item in list_of_items:
            print(item["item"]["image"][0])
            skins[item["item"]["name"]] = item["item"]["image"][0]

        print(f"Successfully scraped {len(skins)} skins for {weapon_name}")
        return skins

    except requests.RequestException as e:
        print(f"Error scraping {weapon_name}: {e}")
        print(f"Response status code: {getattr(e.response, 'status_code', 'N/A')}")
        return {}  # Return empty dict instead of list to be consistent
```

```python
weapons = ["AK-47"]
all_skins = {}

for weapon in weapons:
    print(f"\nScraping skins for {weapon}...")
    skins = scrape_weapon_skins(weapon)
    if skins:
        all_skins[weapon] = skins

    # Add a longer delay between weapons
    time.sleep(random.uniform(2, 3))

if all_skins:
    with open("cs_skins.json", "w", encoding="utf-8") as f:
        json.dump(all_skins, f, indent=2, ensure_ascii=False)
    print("\nData saved to cs_skins.json")
else:
    print("\nNo data was collected. Please check the website's accessibility.")
```
